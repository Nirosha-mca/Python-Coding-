```python
# Task 1 — Data Import & Setup
# Import CSV using Pandas
# Check/convert data types
# Convert date columns to datetime
df['posted_at'] = pd.to_datetime(df['posted_at'],format='%d-%m-%Y')
df['posted_at'] = pd.to_datetime(df['posted_at'],format = '%H-%M-%S')
df
```

```python
# Task 2 — Data Cleaning
# Cleaning Missing Data
# Detect missing values (isnull(), isna())
# Handle using: dropna(), fillna(), median/mode, forward/backward-fill
df.isnull()
```

```python
df.isna().sum()
```

```python
# Handle using: dropna(), fillna(), median/mode, forward/backward-fill
df['likes'] = df['likes'].fillna(df['likes'].median())
df['age'] = df['age'].fillna(df['age'].median())
df['comments'] = df['comments'].fillna(df['comments'].median())
df['shares'] = df['shares'].fillna(df['shares'].median())
df.isna().sum()
```

```python
# imputing mode --- own categorical mode
df['gender'] = df['gender'].fillna("Other")
df['sentiment'] = df['sentiment'].fillna("unknown")
df.isnull().sum()
df
```

```python
# deleting particular columns by using dropna()
df.dropna(subset='user_id')
df
```

```python
# Duplicate Handling- Identify & remove duplicates
df.duplicated('user_id').sum()
```

```python
df.drop_duplicates('user_id')
```

```python
# Data Formatting
# Fix incorrect data types
# Standardize categories (e.g., gender labels)
# Correct unrealistic values in likes, comments, shares
df['age'] = df['age'].astype(int)
df['likes'] = df['likes'].astype(int)
df['comments'] = df['comments'].astype(int)
df['shares'] = df['shares'].astype(int)
df
```

```python
df.dtypes
```

```python
# Standardize categories (e.g., gender labels)
df['gender'] = df['gender'].replace({'m':'Male','f':'Female','O':'Other'})
df
```

```python
# Feature Cleaning
# Extract hashtag count
# Clean sentiment labels
df['hashtags'] = pd.DataFrame({'caption':['#fun','#happy','#enjoyment','#foodie','#amazing day']
})
df['hashtag_count'] = df['hashtags'].str.count('#')
df
```

```python
# Clean sentiment labels
df['sentiment'] = df['sentiment'].replace({'pos':'Positive','neg':'Negative','neu':'Neutral'})
df
```
